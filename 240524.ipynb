데이터의 기본 정보

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.info()
```

데이터의 기초 통계량

```python
df.describe()
```

```python
# 품종별 개수
value_counts = df['species'].value_counts()
value_counts
```

```python
# 결측치 확인
df.isnull().sum()
df.isna().sum()
```

```python
# 중복 데이터 확인
df[df.duplicated()]
```

```python
# 중복 데이터 모두 확인
idx = (df['sepal_length'] == 5.8) & (df['petal_width'] == 1.9)
df.loc[idx,]
```

```python
# 중복 데이터 삭제
df2 = df.drop_duplicates()
```

```python
# 품종 열을 기준으로 합계 구하기
df2.groupby('species').sum()
```

```python
# 품종 열을 기준으로 평균 구하기
df2.groupby('species').mean()
```

```python
# 막대 그래프 그리기
import matplotlib.pyplot as plt
plt.bar(df.index[:10], df.sepal_length[:10])
plt.title('Sepal Length Visualization')
plt.xlabel('Data')
plt.ylabel('Sepal length (cm)')
plt.show()
```

```python
df2 = df.groupby('species').mean()
plt.bar(df2.index, df2['sepal_length'])
plt.title('species sepal length graph')
plt.xlabel('species')
plt.ylabel('sepal_lengths mean')
plt.show()
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

df[features].hist(bins=20, figsize=(10, 8), edgecolor='black')
plt.suptitle('Histograms of Iris Dataset Features', size=16)
plt.show()
```

```python
import seaborn as sns
iris_data=df
sns.scatterplot(x='sepal_length', y='petal_length', data = iris_data)
plt.title('Scatter Plot', fontsize=20)
plt.show()
```
