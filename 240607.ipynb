```python
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())
train_data['Embarked'] = train_data['Embarked'].fillna(train_data['Embarked'].mode()[0])
train_data['Fare'] = train_data['Fare'].fillna(train_data['Fare'].median())
```

```python
train_data.isnull().sum()
```

```python
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median()) #결측치_제거
test_data['Embarked'] = test_data['Embarked'].fillna(test_data['Embarked'].mode()[0])
test_data['Fare'] = test_data['Fare'].fillna(test_data['Fare'].median())
```

```python
test_data.isnull().sum()
```

```python
train_data = pd.get_dummies(train_data, columns = ['Sex', 'Embarked']) #하나의_컬럼을_나눠_두개의_컬럼으로_생성
```

```python
test_data = pd.get_dummies(test_data, columns = ['Sex', 'Embarked'])
```

```python
train_data.head()
```

```python
train_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1, inplace = True) #필요없는_데이터_제거
```

```python
test_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = train_data.drop('Survived', axis = 1)
y = train_data['Survived']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2)
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state = 42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(y_val, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
conf_matrix = confusion_matrix(y_val, y_pred)
print('Confusion_matrix : ')
print(conf_matrix)
```

```python
class_report = classification_report(y_val, y_pred)
print('Classification_report : ')
print(class_report)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (20, 10))
plot_tree(clf, filled = True, feature_names = X.columns, class_names = ['Not Survived', 'Survived'])
plt.show()
```
